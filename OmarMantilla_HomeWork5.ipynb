```r
# Loading libraries
library('ISLR')
library('e1071')

# Loading Dataset
attach(OJ)
```

```r
# A

set.seed(1)
train = sample(dim(OJ)[1], 800)
train_df = OJ[train, ]
test_df = OJ[-train, ]
```

```r
train_df
```

```r
# B
yu=svm(Purchase~ .,kernel="linear",data=train_df,cost=0.01)
summary(yu)
```

Number of Support Vectors:  435

CH = 219
MM = 216

```r
# C
pred = predict(yu, train_df)
pred
```

```r
table(train_df$Purchase, pred)
```

```r
(75 + 65)/(75 + 65+420  +420  )
```

```r
predT = predict(yu, test_df)
table(test_df$Purchase, predT)
```

```r
(33+15)/(153+69+33+15)
```

```r
# D
set.seed(1)
out=tune(svm,Purchase~.,data=train_df,kernel="linear",ranges=list(cost=10^seq(-2,1,by=0.25)))
summary(out)
```

```r
# E
yutunned=svm(Purchase~.,kernel="linear",data=train_df,cost=out$best.parameters$cost)
pred2 = predict(yutunned, train_df)
table(train_df$Purchase, pred2)
```

```r
(70+62)/(70+62+423+245)
```

```r
predT2 = predict(yutunned, test_df)
table(test_df$Purchase, predT2)
```
